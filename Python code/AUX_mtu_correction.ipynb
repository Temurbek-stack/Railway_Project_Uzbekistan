```python
import geopandas as gpd
from shapely.geometry import Point, LineString
from shapely.ops import split
import pandas as pd
import numpy as np

# Reading shapefiles
buxoro_mtu = gpd.read_file("/content/drive/MyDrive/UTY/UTY/3. qgis project/railway_tracks_and_stations_cleaned/MTUs/buxoro_mtu_tracks_shp/buxoro_mtu_tracks_shp.shp")
qungirat_mtu = gpd.read_file("/content/drive/MyDrive/UTY/UTY/3. qgis project/railway_tracks_and_stations_cleaned/MTUs/qungirat_mtu_tracks_shp/qungirat_mtu_tracks_shp.shp")
qar = gpd.read_file("/content/drive/MyDrive/UTY/UTY/3. qgis project/railway_tracks_and_stations_cleaned/MTUs/qarshi_mtu_tracks_shp/qarshi_mtu_tracks_shp.shp")
ter = gpd.read_file("/content/drive/MyDrive/UTY/UTY/3. qgis project/railway_tracks_and_stations_cleaned/MTUs/termiz_mtu_tracks_shp/termiz_mtu_tracks_shp.shp")
tash = gpd.read_file("/content/drive/MyDrive/UTY/UTY/3. qgis project/railway_tracks_and_stations_cleaned/MTUs/tashkent_mtu_tracks_shp/tashkent_mtu_tracks_shp.shp")
qoq = gpd.read_file("/content/drive/MyDrive/UTY/UTY/3. qgis project/railway_tracks_and_stations_cleaned/MTUs/qoqon_mtu_tracks_shp/qoqon_mtu_tracks_shp.shp")
```

```python
# Adding MTU column
qungirat_mtu['MTU'] = 'Kungirat'
buxoro_mtu['MTU'] = 'Bukhara'
qar['MTU'] = 'Karshi'
ter['MTU'] = 'Termez'
qoq['MTU'] = 'Kokand'
tash['MTU'] = 'Tashkent'

# Combining all into one GeoDataFrame
railways_by_MTUs = gpd.GeoDataFrame(pd.concat([qungirat_mtu, buxoro_mtu, qar, ter, tash, qoq], ignore_index=True))
```

```python
railways_by_MTUs = railways_by_MTUs.to_crs('EPSG:4326')
# Check for missing geometries
missing_geometries = railways_by_MTUs['geometry'].isna()
if missing_geometries.any():
    print("Warning: Some rows have missing geometries.")

```
